```python
import pandas as pd
```

```python
arquivo_cimehgo = r"C:\Users\thiag\OneDrive\Área de Trabalho\TCC 2\Scripts\Arquivos Processados\CIMEHGO_TRATADO.xlsx"
arquivo_estacoes = r"C:\Users\thiag\OneDrive\Área de Trabalho\TCC 2\Scripts\Arquivos Finais\Locais de Coleta.xlsx"
saida_final = r"C:\Users\thiag\OneDrive\Área de Trabalho\TCC 2\Scripts\Arquivos Processados\CIMEHGO_FINAL.xlsx"
```

```python
df_cimehgo = pd.read_excel(arquivo_cimehgo)
df_estacoes = pd.read_excel(arquivo_estacoes)

df_cimehgo.columns = [c.strip().upper() for c in df_cimehgo.columns]
df_estacoes.columns = [c.strip().upper() for c in df_estacoes.columns]
```

```python
if 'ESTAÇÃO' not in df_cimehgo.columns:
    raise ValueError("CIMEHGO_TRATADO não tem coluna 'ESTAÇÃO'")
if 'ESTAÇÃO' not in df_estacoes.columns:
    raise ValueError("Locais de Coleta não tem coluna 'ESTAÇÃO'")

df_final = df_cimehgo.merge(df_estacoes[['ESTAÇÃO', 'LAT', 'LONG']],
                            on='ESTAÇÃO',
                            how='left')

df_final = df_final.rename(columns={'LAT': 'LATITUDE', 'LONG': 'LONGITUDE'})

df_final.to_excel(saida_final, index=False)
```
